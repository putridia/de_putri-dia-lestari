### 1. Mengelola Nilai yang Hilang pada kolom area dan floor

```python
import pandas as pd

df = pd.read_csv('house_listings.csv')

# Menghitung jumlah nilai yang hilang pada kolom 'area'
num_missing_area = df['area'].isna().sum()

# Menghitung jumlah nilai yang hilang pada kolom 'floor'
num_missing_floor = df['floor'].isna().sum()

# Menampilkan jumlah nilai yang hilang
print(f"Jumlah nilai yang hilang pada kolom 'area': {num_missing_area}")
print(f"Jumlah nilai yang hilang pada kolom 'floor': {num_missing_floor}")
```

```python
df['area'] = df['area'].astype(str)
df['area'] = df['area'].str.replace(' ','')
df['area'] = df['area'].str.replace('m²','')
df['area'] = pd.to_numeric(df['area'], errors='coerce')
mean_area = df['area'].mean()
df['area'].fillna(mean_area, inplace=True)

df['floor'] = df['floor'].astype(str)
df['floor'] = df['floor'].str.replace(' ','')
df['floor'] = df['floor'].str.replace('/','.')
df['floor'] = pd.to_numeric(df['floor'], errors='coerce')
median_floor = df['floor'].median()
df['floor'].fillna(median_floor, inplace=True)

missing_values_after_imputation = df[['area', 'floor']].isnull().sum()
print("\nJumlah nilai yang hilang setelah imputasi:")
print(missing_values_after_imputation)
```

### 2. Mengatasi Outlier

```python
from IPython.display import display

df = pd.read_csv('house_listings.csv')
df_sorted = df.sort_values(by='price')
df_sorted['price'] = pd.to_numeric(df['price'].str.replace(' ', ''), errors='coerce')

Q1 = df_sorted['price'].quantile(0.25)
Q3 = df_sorted['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - (1.5*IQR)
upper_bound = Q3 + (1.5*IQR)

outliers = df_sorted[(df_sorted['price'] < lower_bound) | (df_sorted['price'] > upper_bound)]
df_outliers = pd.DataFrame(outliers)

df_cleaned = df_sorted[(df_sorted['price'] >= lower_bound) & (df_sorted['price'] <= upper_bound)]
df_data_cleaned = pd.DataFrame(df_cleaned)

print("Data asli:")
display(df_sorted)

print("Data yang terdeteksi sebagai outlier:")
display(df_outliers)

print("Data setelah menghapus outlier:")
display(df_data_cleaned)
```
