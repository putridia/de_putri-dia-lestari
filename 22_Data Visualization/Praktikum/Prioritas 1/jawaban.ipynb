# Soal Prioritas 1 no 1

Buatlah sebuah data visualization dari dataset berikut dengan kriteria: [Dataset:](https://gist.github.com/nadirbslmh/7a3d729818dd09bb195ef998033638ce)
   
   a. Membuat sebuah line plot untuk menggambarkan jumlah nilai dari transaksi per hari.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('electronic_shop.csv')

grouped_data = df.groupby('transaction_date')['transaction_amount'].sum().reset_index()

plt.figure(figsize=(15, 7))
plt.plot(grouped_data['transaction_date'], grouped_data['transaction_amount'])
plt.title("Total Transaction Amount per Day")
plt.xlabel("Date")
plt.ylabel("Transaction Amount")

plt.show()
```

    b. Membuat sebuah bar plot untuk menggambarkan jumlah customer per hari.

```python
import seaborn as sns

# Group by transaction_date and count customer_name
daily_customers = df.groupby('transaction_date')['customer_name'].nunique().reset_index()

# Plot the bar plot
plt.figure(figsize=(15, 7))
sns.barplot(x='transaction_date', y='customer_name', data=daily_customers)

plt.title('Daily Number of Customers')
plt.xlabel('Date')
plt.ylabel('Number of Customers')

plt.show()
```

c. Membuat sebuah scatter plot untuk menggambarkan jumlah nilai dari transaksi per hari.

```python
transaction_dates = df['transaction_date'].unique()
daily_transaction_amount = df.groupby('transaction_date')['transaction_amount'].sum()

plt.figure(1, figsize= (15, 7))
plt.scatter(transaction_dates, daily_transaction_amount)

plt.figure(2, figsize= (15, 7))
plt.plot(transaction_dates, daily_transaction_amount)

plt.show()
```

# Soal Prioritas 1 no 2

Sebuah toko roti / bakery telah merekap data transaksi. Toko tersebut ingin mendapatkan informasi berikut: 

a. Jumlah nilai transaksi per hari.

```python
df = pd.read_csv("bakery.csv")

transaction_sum = df.groupby('transaction_date')['transaction_amount'].sum().reset_index()

transaction_sum.plot(x='transaction_date', y='transaction_amount', kind='bar', figsize=(15, 7))
plt.title("Total Transaction Amount per Day")

plt.xlabel("Date")
plt.ylabel("Transaction Amount")

plt.show()
```

b. Jenis roti yang paling banyak dipesan

```python
df = pd.read_csv("bakery.csv")

item_count = df.groupby(['item_name']).count()
category = item_count.index.get_level_values(0).to_list()
transaction_count = item_count['transaction_id']

plt.figure(figsize=(15, 7))
plt.title("Jenis Roti yang Paling Banyak Dipesan")
plt.pie(
    transaction_count,
    labels=category,
    startangle=90,
    # autopct='%1.1f%%'
    # autopct="%.0f%%"
    autopct="%.0f%%"
)

plt.show()
```

c. Rata-rata kepuasan pelanggan (customer_rating) per hari

Berdasarkan poin-poin tersebut, buatlah data visualization untuk memperoleh informasi tersebut.

```python
df = pd.read_csv("bakery.csv")

customer_rating_mean = df.groupby("transaction_date")["customer_rating"].mean()

plt.figure(figsize=(12, 6))
customer_rating_mean.plot(kind="line")

plt.title("Rata-rata Kepuasan Pelanggan Per Hari")
plt.xlabel("Tanggal")
plt.ylabel("Rata-rata Kepuasan Pelanggan")

plt.show()
```
