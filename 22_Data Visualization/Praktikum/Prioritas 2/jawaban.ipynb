Sebuah platform e-commerce telah merekap data transaksi. Platform tersebut ingin mendapatkan informasi berikut: [Dataset](https://gist.githubusercontent.com/nadirbslmh/8fc9cc6cd5cbaaf5cbff63b090fb497e/raw/a7bf3e1edab88b04314a40a9de3ed744bc86d0e9/ecommerce.json)

a. Jumlah nilai transaksi per hari.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json("e-commerce.json")

transaction_dates = df.groupby(['transaction_date']).sum(numeric_only=True)
date = transaction_dates.index.get_level_values(0).to_list()
transaction = transaction_dates['transaction_amount']

plt.figure(figsize=(12, 6))
transaction = df.groupby("transaction_date")["transaction_amount"].sum()

transaction.plot(kind="bar")
plt.title("Jumlah Nilai Transaksi Per Hari")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Nilai Transaksi")
plt.show()
```

b. Kategori barang yang paling banyak dibeli.

```python
df = pd.read_json('e-commerce.json')

transaction_category = df.groupby('item_category').count()
category = transaction_category.index.get_level_values(0).to_list()
amount = transaction_category['transaction_id']

plt.figure(figsize=(15,7))
plt.title("Amount Sold per Category")
plt.xlabel("Category")
plt.ylabel("Amount Sold")
plt.bar(category, amount, color='yellow')

plt.show()
```

c. Jumlah nilai transaksi dengan metode pembayaran e-wallet.

```python
# Menghitung jumlah nilai transaksi dengan metode pembayaran e-wallet
ewallet_transaction_amount = df[df['payment_method'] == 'e wallet']['transaction_amount'].sum()

# Menampilkan jumlah nilai transaksi dengan metode pembayaran e-wallet
print("Jumlah nilai transaksi dengan metode pembayaran e-wallet:", ewallet_transaction_amount)
```

# Soal Prioritas 2 no 2

Sebuah survei mengenai pengembangan software telah dilakukan dan didapatkan data sebagai berikut. Buatlah data visualization untuk memperoleh berbagai informasi berikut:

a. Persentase responden berdasarkan umur.

```python
df = pd.read_csv('survey.csv')

df['age'] = df['age'].replace('-', 'Unknown')

# Kelompokkan data berdasarkan rentang umur dan hitung jumlah responden untuk setiap kelompok
age_count_result = df.groupby(['age']).size()

# Siapkan data untuk visualisasi
age_count = age_count_result.values
age_labels = age_count_result.index

# Buat pie chart
plt.figure(figsize=(5, 5))
plt.pie(
    age_count, 
    labels=age_labels, 
    autopct='%.0f%%',
    startangle=140
)

plt.title('Persentase Responden Berdasarkan Umur')
plt.show()
```

b. Bahasa pemrograman / framework yang paling banyak digunakan dalam pengembangan front end.

```python
df = pd.read_csv("survey.csv")

# Menghitung frekuensi masing-masing bahasa pemrograman/framework front end
front_end_counts = df['favorite_front_end'].value_counts()

# Plotting
plt.figure(figsize=(7,5))
front_end_counts.plot(kind='bar', color='skyblue')
plt.title('Frekuensi Bahasa Pemrograman/Framework Front End yang Digunakan')
plt.xlabel('Bahasa Pemrograman/Framework Front End')
plt.ylabel('Frekuensi')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

c. Bahasa pemrograman / framework yang paling banyak digunakan dalam pengembangan back end.

```python
df = pd.read_csv("survey.csv")

df = df[df['favorite_back_end'] != "-"]

# Hitung frekuensi penggunaan setiap bahasa pemrograman/framework pada back end
backend_count_result = df['favorite_back_end'].value_counts()

# Ambil nama bahasa pemrograman/framework dan frekuensinya
backends = backend_count_result.index.tolist()
backend_count = backend_count_result.values.tolist()

# Buat visualisasi dengan diagram batang
plt.figure(figsize=(10, 6))
plt.bar(backends, backend_count, color='skyblue')
plt.title("Bahasa Pemrograman/Framework yang Paling Banyak Digunakan dalam Pengembangan Back End")
plt.xlabel("Bahasa Pemrograman/Framework")
plt.ylabel("Jumlah Penggunaan")
plt.xticks(rotation=45, ha='right')  # Rotasi label agar lebih mudah dibaca
plt.tight_layout()

# Tampilkan visualisasi
plt.show()
```

d. Persentase bahasa pemrograman / framework yang dipilih dalam pengembangan front end.

```python
df = pd.read_csv('survey.csv')

# Clean and preprocess the data
df = df.drop_duplicates()  # Remove duplicate rows
front_end_categories = ["JavaScript", "React", "HTML/CSS", "Angular", "Vue.js", "Python", "Node.js", "Java", "C#", "PHP"]
df = df[df["favorite_front_end"].isin(front_end_categories)]  # Remove rows with unrecognized front-end categories

# Calculate the percentage of each front-end programming language/framework
front_end_counts = df["favorite_front_end"].value_counts()
total_responses = len(df)
front_end_percentages = front_end_counts / total_responses * 100

# Create a pie chart to visualize the results
plt.pie(
    front_end_percentages,
    labels=front_end_counts.index,
    autopct="%.0f%%"
)

plt.title("Front-End Programming Languages/Frameworks Preference")
plt.axis("equal")
plt.show()
```

e. Rentangan gaji per tahun yang paling banyak berdasarkan hasil survei.

```python
df = pd.read_csv('survey.csv')

survey_by_salary = df.groupby(['salary_per_year_in_USD']).count()

salary_range = survey_by_salary.index.get_level_values(0).to_list()
count = survey_by_salary['name']

# Membuat visualisasi menggunakan bar chart
plt.figure(figsize=(10, 6))
plt.bar(salary_range, count, color='skyblue')

plt.xlabel('Rentang Gaji per Tahun')
plt.ylabel('Jumlah Responden')
plt.title('Distribusi Responden Berdasarkan Rentang Gaji per Tahun')

plt.xticks(rotation=45, ha='right')  # Memutar label sumbu x untuk memastikan kejelasan
plt.tight_layout()  # Mengatur layout agar tidak tumpang tindih
plt.show()
```
